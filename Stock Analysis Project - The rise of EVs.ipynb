# Daily Percentage Change

 $ R_{t} = \frac{P_{r}}{P_{r-1}}- 1 $

```python
ford['returns'] = (ford['Close']/ford['Close'].shift(1)) - 1
nio['returns'] = (nio['Close']/nio['Close'].shift(1)) - 1
xpeng['returns'] = (xpeng['Close']/xpeng['Close'].shift(1)) - 1
```

```python
ford['returns'].hist(bins=100, label = "Ford", alpha = 1, figsize = (13,6))
nio['returns'].hist(bins=100, label = "Nio", alpha = .6)
xpeng['returns'].hist(bins=100, label = "Xpeng", alpha = .6)
plt.legend()
```

```python
ford['returns'].plot(kind='kde', label = 'Ford', figsize = (15,6))
nio['returns'].plot(kind='kde', label = 'Nio')
xpeng['returns'].plot(kind='kde', label = 'Xpeng')
```

```python
box_df = pd.concat([ford['returns'],nio['returns'],xpeng['returns']],axis = 1)
box_df.columns = ['Ford Returns', 'Nio Returns', 'Xpeng Returns']
box_df.plot(kind='box')
```

```python
scatter_matrix(box_df,figsize=(8,8),hist_kwds={'bins':50},alpha = 0.5)
```

# Cumulative Return

$ i_{t} = (1+r_{t})i_{t-1} $

```python
#Reports the gain/loss made from if invested at the start of Sep 2020 to Sep 2021

ford['Cumulative Return'] = (1 + ford['returns']).cumprod()
nio['Cumulative Return'] = (1 + nio['returns']).cumprod()
xpeng['Cumulative Return'] = (1 + xpeng['returns']).cumprod()
```

```python
ford['Cumulative Return'].plot(label='Ford', figsize = (15,7))
nio['Cumulative Return'].plot(label='Nio', figsize = (15,7))
xpeng['Cumulative Return'].plot(label='Xpeng', figsize = (15,7))
plt.title('Cumulative Return vs Time', size = 15)
plt.legend()
plt.ylabel("Cumulative Return (100%)")

#Example, if invested $1 dollar in September 2020, if invested in NIO you would get a cumulative return of 230%
```
